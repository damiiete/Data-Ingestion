```python
from github import Github
import pandas as pd
import config
```

```python
#github account access token
API_KEY = config.API_KEY

#initialize github client
g = Github(API_KEY)
```

```python
def get_user_info(query):
    
    #dictionary to hold user info
    user_dict = {'Name': [], 'Login': [], 'Bio': [], 'Blog': [], 'Company': [], 'Email': [],
                 'Followers(count of followers)': [], 'Following(count of following)': [], 'Id': [], 'Url': [], 'Location': []}
                
    users = g.search_users(query)

    #populate the user dictionary
    for user in users:
      user_dict['Name'].append(user.name)
      user_dict['Login'].append(user.login)
      user_dict['Bio'].append(user.bio)
      user_dict['Company'].append(user.company)
      user_dict['Email'].append(user.email)
      user_dict['Followers(count of followers)'].append(user.followers)
      user_dict['Following(count of following)'].append(user.following)
      user_dict['Id'].append(user.id)
      user_dict['Blog'].append(user.blog)
      user_dict['Url'].append(user.url)
      user_dict['Location'].append(user.location)
  
    return user_dict
```

```python
#to get query from user input
def get_query():
  query = input('Ex: Machine Learning location:Lagos\n')
  return query
```

```python
#get query from user input
query = get_query()
```

```python
#get user info in a dictionary
user_dict = get_user_info(query)
```

```python
#create dataframe from dictionary
ML_users_df = pd.DataFrame(user_dict)


print(ML_users_df.info())
ML_users_df.head()

```
